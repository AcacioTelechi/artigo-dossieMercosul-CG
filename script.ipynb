```python
import pandas as pd
```

```python
base = pd.read_excel('05-10.xlsx')
```

```python
def get_list_unique(lista):
    lista_final = []
    for a in lista:
        if type(a) is str:
            if ";" in a:
                b = a.split(';')
                for c in b:
                    p = c.strip()
                    if p not in lista_final:
                        lista_final.append(p)
            else:
                p = a.strip()
                if p not in lista_final:
                    lista_final.append(p)
    return lista_final
```

```python
# quantidade de palavras-chave
kws = [k for k in base['Author Keywords']]
kws_final = get_list_unique(kws)
len(kws_final)
```

```python
# quantidade de autores
autores = [autor for autor in base['Authors']]
autores_final = get_list_unique(autores)
len(autores_final)
```
